```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import os
os.listdir()
```

```python
df.dropna(inplace=True)
```

```python
top = df.groupby("Item_Type")["Item_Outlet_Sales"].sum().sort_values(ascending=False)
```
